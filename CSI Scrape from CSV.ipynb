```python
directory = "/Users/subodh.sonar/Desktop/CSI/SS_Runs_03092022"
length = 0
paths = []
for root, subdirectories, files in os.walk(directory):
    for subdirectory in subdirectories:
        for file in files:
             if fnmatch.fnmatch(file, '*strip_analysis_data_*'):
                temp = os.path.join(root,file)
                paths.append(temp)
                length=length+1
print(length)                
```

```python
out_list = []
for i in range(len(paths)):

    df = pd.DataFrame()
    df = pd.read_csv(paths[i], skiprows=8)
    for index, row in df.iterrows():
    
        out_list.append((row["Image Name"],row["Total Volume"], row["Bead Volume"], row["GB/Tot"]))
```
